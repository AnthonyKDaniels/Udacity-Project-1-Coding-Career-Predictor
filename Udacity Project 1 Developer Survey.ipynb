We have 114 columns/features in our dataset.  We need to start the cleaning process.
Our project is going to attempt create a model that predicts those who will create a career which will be defined as those with over 10 years work experience.

```python
# eliminate columns that we don't want to be taken into account for our model
# we have a column 'check' that appears to have responses all 'apples'.  check to see if there are any other responses

# Check for non-'apples' responses in the 'check' column
non_apples = survey_responses[survey_responses['Check'] != 'Apples']

# Display results
if len(non_apples) > 0:
    print(f"Found {len(non_apples)} rows with non-'apples' responses:")
    print(non_apples[['Check']])  # Show only the 'check' column
else:
    print("All responses in 'Check' column are 'Apples'.")
```

Based on the fact that all of the responses are 'Apples' under the 'Check' column, we can assume that it is a nonimpactful column since it applies to everybody in the survey

```python
#code to remove undesired columns from file
survey_responses.drop(columns=['Check', 'OrgSize', 'PurchaseInfluence', 'BuyNewTool', 'BuildvsBuy', 'TechEndorse', 'Currency','CompTotal', 'LanguageAdmired', 'DatabaseAdmired', 'PlatformAdmired', 'WebframeAdmired', 'EmbeddedAdmired', 'MiscTechWantToWorkWith', 'MiscTechAdmired', 'ToolsTechAdmired', 'NEWCollabToolsWantToWorkWith', 'NEWCollabToolsAdmired', 'OfficeStackAsyncWantToWorkWith', 'OfficeStackAsyncAdmired', 'OfficeStackSyncWantToWorkWith', 'OfficeStackSyncAdmired', 'AISearchDevWantToWorkWith', 'AISearchDevAdmired', 'SurveyLength', 'SurveyEase', 'ConvertedCompYearly'], inplace=True)
```

```python
WorkExp_NA = survey_responses['WorkExp'].isna().sum()  #determine if there are NA to be dealth with in WorkExp column
print (WorkExp_NA)
```

```python
survey_responses['Career'] = survey_responses['WorkExp'].fillna(0) >=10   #creates new column named Career and assigns False to everything w/NA or Less than 10 years of work experience
```
