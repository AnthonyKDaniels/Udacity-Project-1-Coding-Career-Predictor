```python
survey_responses['WorkExp'] = survey_responses['WorkExp'].fillna(0)  #replaces all 'NaN' rows under 'WorkExp' to an assumed work experience of 0
```

```python
survey_responses['MainBranch'] = survey_responses['MainBranch'].fillna('No_Experience') ### Replaces all 'Nan' responses in 'MainBranch' column to 'No_Experience'
```

```python
###Converting Responses under 'MainBranch' column to individual columns with Boolean responses

Main_Branch_choices = [
    'I am a developer by profession',
    'I am not primarily a developer, but I write code sometimes as part of my work/studies',
    'I used to be a developer by profession, but no longer am',
    'I am learning to code',
    'I code primarily as a hobby',
    'No_Experience'
]

# Create boolean columns correctly
for option in Main_Branch_choices:
    # Improved column name cleaning
    col_name = (
        option.lower()  # Convert to lowercase
        .replace(' ', '_')
        .replace('/', '_')
        .replace(',', '')
        .replace("'", "")  # Remove apostrophes
    )
    # Compare against the current option only (not the whole list)
    survey_responses[f"Branch_{col_name}"] = (survey_responses['MainBranch'] == option)
```

```python
## Deal w/age object column which has ranges vs actual ages.  convert to integer ages in new 'age_numeric' column
## To avoid too many columns, will use median age points to create integer based row

# Create age mapping dictionary (using midpoint of each range)
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,  # (18+24)/2
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None  
}

survey_responses['age_numeric'] = survey_responses['Age'].map(age_mapping)
```

```python
# Deal w/No Responses or NaN data points in newly created 'age_numeric' column
median_age = survey_responses['age_numeric'].median()
survey_responses['age_numeric'] = survey_responses['age_numeric'].fillna(median_age)
```

Deal with EdLevel Column

```python
# Define education level options
ed_levels = [
    'Primary/elementary school',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)',
    'Some college/university study without earning a degree',
    'Associate degree (A.A., A.S., etc.)',
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)',
    'Something else'
]

# Create boolean columns
for level in ed_levels:
    # Generate clean column name
    col_name = (
        level.lower()
        .replace(' ', '_')
        .replace('/', '_')
        .replace('(', '')
        .replace(')', '')
        .replace('.', '')
        .replace(',', '')
        .replace('’', '')
        .replace('etc', '')
        .replace('e.g.', '')
        .strip('_')
    )
    
    # Create boolean column
    survey_responses[f'ed_{col_name}'] = survey_responses['EdLevel'] == level

# Deal with missing values or "NA's"
survey_responses['ed_missing'] = survey_responses['EdLevel'].isna()

```

Deal with column 'LearnCode'

```python


# Define learning methods (from your list)
learn_methods = [
    'Books / Physical media',
    'Coding Bootcamp',
    'Colleague',
    'Friend or family member',
    'Online Courses or Certification',
    'On the job training',
    'Other online resources (e.g., videos, blogs, forum, online community)',
    'School (i.e., University, College, etc)',
    'Other (please specify)'
]

# Clean method names for column creation
def clean_method_name(method):
    return (
        method.lower()
        .replace(' ', '_')
        .replace('/', '_')
        .replace('(', '')
        .replace(')', '')
        .replace('.', '')
        .replace(',', '')
        .replace('e.g.', '')
        .replace('i.e.', '')
        .strip('_')
    )

# Standardize delimiters and split responses
def split_learn_code(responses):
    if pd.isna(responses):
        return []
    # Replace semicolons with commas, then split
    return [x.strip() for x in responses.replace(';', ',').split(',')]

# Create boolean columns
for method in learn_methods:
    col_name = f"learn_{clean_method_name(method)}"
    survey_responses[col_name] = (
        survey_responses['LearnCode']
        .apply(lambda x: method in split_learn_code(x))
    )

# Handle NA values
survey_responses['learn_missing'] = survey_responses['LearnCode'].isna()



```
