Deal with column 'LearnCode'

```python


# Define learning methods (from your list)
learn_methods = [
    'Books / Physical media',
    'Coding Bootcamp',
    'Colleague',
    'Friend or family member',
    'Online Courses or Certification',
    'On the job training',
    'Other online resources (e.g., videos, blogs, forum, online community)',
    'School (i.e., University, College, etc)',
    'Other (please specify)'
]

# Clean method names for column creation
def clean_method_name(method):
    return (
        method.lower()
        .replace(' ', '_')
        .replace('/', '_')
        .replace('(', '')
        .replace(')', '')
        .replace('.', '')
        .replace(',', '')
        .replace('e.g.', '')
        .replace('i.e.', '')
        .strip('_')
    )

# Standardize delimiters and split responses
def split_learn_code(responses):
    if pd.isna(responses):
        return []
    # Replace semicolons with commas, then split
    return [x.strip() for x in responses.replace(';', ',').split(',')]

# Create boolean columns
for method in learn_methods:
    col_name = f"learn_{clean_method_name(method)}"
    survey_responses[col_name] = (
        survey_responses['LearnCode']
        .apply(lambda x: method in split_learn_code(x))
    )

# Handle NA values
survey_responses['learn_missing'] = survey_responses['LearnCode'].isna()



```

Deal with 'YearsCode' and 'YearsCodPro' columns.  Converting from object to integer columns and dealing with string responses

```python
import numpy as np

def convert_years_column(series):
    """
    Converts a years column with mixed responses to integers.
    Handles:
    - 'Less than 1 year' → 0
    - 'More than 50 years' → 51
    - Numeric strings → integers
    - NA values → np.nan
    """
    # Convert special cases first
    series = series.replace({
        'Less than 1 year': '0',
        'More than 50 years': '51'
    })
    
    # Convert to numeric, coercing errors to NaN
    return pd.to_numeric(series, errors='coerce')

# Apply to both columns
survey_responses['YearsCode'] = convert_years_column(survey_responses['YearsCode'])
survey_responses['YearsCodePro'] = convert_years_column(survey_responses['YearsCodePro'])

# Create flags for converted values
survey_responses['YearsCode_was_estimated'] = (
    survey_responses['YearsCode'].isna() & 
    survey_responses['YearsCode'].notna()
)
survey_responses['YearsCodePro_was_estimated'] = (
    survey_responses['YearsCodePro'].isna() & 
    survey_responses['YearsCodePro'].notna()
)

# Fill NA w median
code_median = survey_responses['YearsCode'].median()
codepro_median = survey_responses['YearsCodePro'].median()

survey_responses['YearsCode'] = survey_responses['YearsCode'].fillna(code_median)
survey_responses['YearsCodePro'] = survey_responses['YearsCodePro'].fillna(codepro_median)

# Convert to integer
survey_responses['YearsCode'] = survey_responses['YearsCode'].astype('Int64')  # Capital I allows NA
survey_responses['YearsCodePro'] = survey_responses['YearsCodePro'].astype('Int64')

# Verify results by counting responses and determining how many of each number of years there are
print("YearsCode value counts:")
print(survey_responses['YearsCode'].value_counts(dropna=False).head(51))
print("\nYearsCodePro value counts:")
print(survey_responses['YearsCodePro'].value_counts(dropna=False).head(51))
```
